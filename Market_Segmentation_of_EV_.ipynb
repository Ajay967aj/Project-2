```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define price bins and labels
bins = [0, 30000, 60000, 90000, 120000, 150000, 200000, df['PriceEuro'].max()]
labels = ['0-30K', '30K-60K', '60K-90K', '90K-120K', '120K-150K','150K-200K','200K+']

# Create a new column for price categories
df['Price Range'] = pd.cut(df['PriceEuro'], bins=bins, labels=labels)

# Count cars in each price range
df_price_count = df['Price Range'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(10, 5))
ax = sns.barplot(x=df_price_count.index, y=df_price_count.values, palette='viridis')

plt.xlabel('Price Range')
plt.ylabel('Number of Cars Sold')
plt.title('Car Sales by Price Range')

# Add labels on bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
# Fill missing values in FastCharge_KmH with the median (since it's a numerical feature)
df["FastCharge_KmH"].fillna(df["FastCharge_KmH"].median(), inplace=True)

# Encode categorical features using one-hot encoding
df_encoded = pd.get_dummies(df, columns=["PowerTrain", "PlugType", "BodyStyle"])

# Ordinal Encoding for Segment
segment_order = [['A', 'B', 'C', 'D', 'E', 'F', 'N','S']]
oe = OrdinalEncoder(categories=segment_order)
df_encoded['Segment_Encoded'] = oe.fit_transform(df[['Segment']])

# Drop non-relevant columns for clustering
df_encoded.drop(["Brand", "Model","Segment",'Price Range'], axis=1, inplace=True)

# Display cleaned dataset structure
df_encoded.info()
```

```python
numeric_df = df_encoded.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix (Numeric Columns Only)")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded)
```

```python
df_scaled
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42,n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
optimal_k = 9  # Example, adjust based on the Elbow Method result
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_encoded['Cluster'] = kmeans.fit_predict(df_scaled)
```

```python
import seaborn as sns

# Count clusters
sns.countplot(x=df_encoded['Cluster'])
plt.title("Cluster Distribution")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(df_scaled, df_encoded['Cluster'])
print(f'Silhouette Score: {score:.2f}')
```

```python
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Create DataFrame
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df_encoded['Cluster']

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_pca['PC1'], y=df_pca['PC2'], hue=df_pca['Cluster'], palette='viridis')
plt.title("Clusters Visualization (PCA)")
plt.show()
```

```python
from scipy import stats

# Calculate Z-scores
z_scores = stats.zscore(df_scaled)

# Set a threshold for outliers (e.g., Z-score > 3)
outliers = (abs(z_scores) > 3).all(axis=1)

# Remove outliers from the dataset
df_clean = df_encoded[~outliers]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

best_k = 0
best_score = -1

# Test different k values and find the best silhouette score
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    score = silhouette_score(df_scaled, kmeans.labels_)

    if score > best_score:
        best_score = score
        best_k = k

print(f'Optimal k: {best_k} with Silhouette Score: {best_score:.2f}')
```
